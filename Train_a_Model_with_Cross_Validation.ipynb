```python
import kagglehub
import pandas as pd
import os
from sklearn.model_selection import train_test_split, cross_validate
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier

# --- 1. SETUP & DATA LOADING ---
print("Step 1: Downloading and loading data...")
path = kagglehub.dataset_download("mukundsavaliya/california-housing-data")
csv_file = [f for f in os.listdir(path) if f.endswith('.csv')][0]
df = pd.read_csv(os.path.join(path, csv_file))

# --- 2. PREPROCESSING (Required for Task 3) ---
print("Step 2: Preprocessing for Classification...")
# Handle missing values and Categorical encoding
df_encoded = pd.get_dummies(df)
imputer = SimpleImputer(strategy='median')
df_imputed = pd.DataFrame(imputer.fit_transform(df_encoded), columns=df_encoded.columns)

# Define Features (X)
X = df_imputed.drop('median_house_value', axis=1)

# Define Target (y) as Classification: 1 if price > median, else 0
# This allows us to calculate Precision and Recall
median_price = df_imputed['median_house_value'].median()
y_class = (df_imputed['median_house_value'] > median_price).astype(int)

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_class, test_size=0.2, random_state=42)

# --- 3. TASK 3: TRAIN WITH CROSS-VALIDATION ---
print("Step 3: Running K-Fold Cross-Validation...")
# Using a Classifier to get requested metrics
model = RandomForestClassifier(n_estimators=100, random_state=42)

# Define the specific metrics requested
scoring = ['accuracy', 'precision', 'recall']

# Perform 5-Fold Cross-Validation
results = cross_validate(model, X_train, y_train, cv=5, scoring=scoring)

# --- FINAL OUTPUT ---
print("\n" + "="*45)
print("     TASK 3 FINAL PERFORMANCE METRICS")
print("="*45)
print(f"Average Accuracy:  {results['test_accuracy'].mean():.4f}")
print(f"Average Precision: {results['test_precision'].mean():.4f}")
print(f"Average Recall:    {results['test_recall'].mean():.4f}")
print("="*45)
print("Interpretation:")
print("- Accuracy: Overall correctness of the model.")
print("- Precision: Quality of 'Expensive' predictions (low false positives).")
print("- Recall: Ability to find all 'Expensive' houses (low false negatives).")
```
