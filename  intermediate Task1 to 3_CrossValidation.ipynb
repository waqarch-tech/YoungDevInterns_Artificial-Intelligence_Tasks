```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler, OneHotEncoder # Fixed spelling here
from sklearn.model_selection import train_test_split

# 1. Prepare the Iris Data
data = load_iris()
X = data.data
y = data.target.reshape(-1, 1)

# Standardize the features (Neural Networks love values centered around 0)
scaler = StandardScaler()
X = scaler.fit_transform(X)

# One-Hot encoding (turning labels into binary vectors)
encoder = OneHotEncoder(sparse_output=False)
y = encoder.fit_transform(y)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# 2. Define the Neural Network
model = Sequential([
    Dense(10, activation='relu', input_shape=(4,)),
    Dense(10, activation='relu'),
    Dense(3, activation='softmax')
])

# 3. Compile and Train
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=50, batch_size=5, verbose=0) # Set verbose to 1 to see progress

# 4. Evaluate
loss, accuracy = model.evaluate(X_test, y_test)
print(f"\n✅ Neural Network Accuracy: {accuracy * 100:.2f}%")
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, LabelEncoder

# 1. CREATE MESSY DATA (Simulating real-world data issues)
# We include missing values (NaN) and text categories
raw_data = {
    'Age': [25, 30, np.nan, 40, 22, 35, np.nan, 28, 45, 32],
    'Salary': [50000, 54000, 60000, 62000, np.nan, 58000, 52000, np.nan, 75000, 61000],
    'Purchased': ['Yes', 'No', 'Yes', 'No', 'Yes', 'No', 'Yes', 'No', 'Yes', 'No']
}

df = pd.DataFrame(raw_data)
print("--- Step 1: Original Messy Data ---")
print(df)

# 2. HANDLE MISSING VALUES (Imputation)
# We replace NaN values with the mean (average) of the column
imputer = SimpleImputer(strategy='mean')
df[['Age', 'Salary']] = imputer.fit_transform(df[['Age', 'Salary']])

# 3. ENCODE CATEGORICAL VARIABLES
# Turning 'Yes'/'No' into 1 and 0
le = LabelEncoder()
df['Purchased'] = le.fit_transform(df['Purchased'])

print("\n--- Step 2 & 3: Cleaned and Encoded Data ---")
print(df)

# 4. NORMALIZE NUMERICAL FEATURES (Scaling)
# This puts Age and Salary on a similar scale (0 to 1 range or mean 0)
scaler = StandardScaler()
X = df.drop('Purchased', axis=1) # Features
y = df['Purchased']             # Target

X_scaled = scaler.fit_transform(X)

# 5. SPLIT DATA INTO TRAINING AND TEST SETS
# 80% for training, 20% for testing
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

print("\n--- Step 4 & 5: Final Preprocessed Shapes ---")
print(f"Training Features Shape: {X_train.shape}")
print(f"Testing Features Shape: {X_test.shape}")
print("\nFirst 3 rows of processed Features (X_train):")
print(X_train[:3])
```

```python
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_validate, KFold

# 1. Load Dataset
# We use the breast cancer dataset as a binary classification example
data = load_breast_cancer()
X, y = data.data, data.target

# 2. Initialize the Model
# Random Forest is robust and handles non-linear relationships well
model = RandomForestClassifier(n_estimators=100, random_state=42)

# 3. Define the Cross-Validation Strategy
# n_splits=5 means the data is divided into 5 chunks (folds)
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# 4. Define the metrics to evaluate
# Using a list allows cross_validate to calculate multiple metrics at once
scoring = ['accuracy', 'precision', 'recall', 'f1']

# 5. Perform Cross-Validation
results = cross_validate(model, X, y, cv=kf, scoring=scoring)

# 6. Display Performance Metrics
print(f"--- Performance Metrics (5-Fold CV) ---")
print(f"Accuracy:  {np.mean(results['test_accuracy']):.4f} (+/- {np.std(results['test_accuracy']):.4f})")
print(f"Precision: {np.mean(results['test_precision']):.4f}")
print(f"Recall:    {np.mean(results['test_recall']):.4f}")
print(f"F1-Score:  {np.mean(results['test_f1']):.4f}")
```
