```python
import kagglehub
import pandas as pd
import os

# Download the dataset
path = kagglehub.dataset_download("mukundsavaliya/california-housing-data")

# List all files in the folder to find the CSV
files = os.listdir(path)
print(f"Files found: {files}")

# Automatically pick the first CSV file found
csv_file = [f for f in files if f.endswith('.csv')][0]
df = pd.read_csv(os.path.join(path, csv_file))

print("Success! Data loaded.")
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer

# 1. Encoding text columns
df_encoded = pd.get_dummies(df)

# 2. Splitting Features (X) and Target (y)
X = df_encoded.drop('median_house_value', axis=1)
y = df_encoded['median_house_value']

# 3. Imputing and Scaling
imputer = SimpleImputer(strategy='median')
scaler = StandardScaler()

X_preprocessed = imputer.fit_transform(X)
X_preprocessed = scaler.fit_transform(X_preprocessed)

# 4. Train/Test Split
X_train, X_test, y_train, y_test = train_test_split(X_preprocessed, y, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_validate

model = RandomForestRegressor(n_estimators=50, random_state=42)

# We use R-squared (accuracy for regression) and MAE (average dollar error)
metrics = ['r2', 'neg_mean_absolute_error']
results = cross_validate(model, X_train, y_train, cv=5, scoring=metrics)

print(f"Average R^2 Score: {results['test_r2'].mean():.3f}")
print(f"Average Error (MAE): ${abs(results['test_neg_mean_absolute_error'].mean()):,.2f}")
```
